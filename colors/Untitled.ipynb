```python
import pandas
import matplotlib.pyplot as plt 
%matplotlib inline

df = pandas.read_csv(  'data.csv' )
df.fillna( -5, inplace=True )

# easy: sequence position 
df['sequence_position'] = df.mutant.str[1:-1].astype( int )

# easy: native residue 
df['native_residue'] = df.mutant.str[0]
df['designed_residue'] = df.mutant.str[-1]

# percent conservation across pfam http://pfam.xfam.org/family/PF00232#tabview=tab3

ala = df[(df.designed_residue == 'A')]
ala = ala.fillna( -4 )

const = ['kcat', '1/km', 'kcat/km']
df[['r', 'g', 'b']] = -1 * ( df[const] - 1 ) / 6 
ala = df[( df.designed_residue == 'A' )]

for name, row in ala.iterrows():
    print 'set_color color{0}=[{3:.2f},0.9,1]; color color{0}, resi {4}; create myobj{0}, resi {4}; set stick_radius, {2:.2f}, myobj{0}; set stick_transparency, {1:.2f}, myobj{0}'.format( row.name, row.r, row.b, row.g, row.sequence_position )
```
